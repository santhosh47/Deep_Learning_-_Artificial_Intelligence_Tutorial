## **Learning Rate**

Recall the update equation that we defined during gradient descent we didn't talk much about this parameter. It's called the **learning rates** and in practice it **determines a lot about how much step we take** and how much trust we take in our gradients.

6S191_MIT_DeepLearning_L1_page-0070.jpg

**Small Learning Rate:**

If we set our learning rate to be very slow then we're model we're having a model that may get stuck in local minima.

Since we're only taking **small steps** towards our gradient, we're going to **converge very slowly**. We may even get stuck if it's too small.

**Large Learning Rate:**

If the learning rate is **too large we might** **overshoot** and actually diverge and our training may kind of explode.

It's not a stable training process. 

So in reality we want to use learning rates that are neither not small not too small not too large to avoid these local Minima and still converge right so we want to kind of use medium-sized learning rates.



## **How to Determine Optimal Learning Rate?**

How do we actually find our learning rate well? 

You have to actually input your learning rate 

1. One way to do it is you could try a bunch of different learning rates and see which one works the best that's actually a very common technique in practice even though it sounds very unsatisfying.

2. Another idea is maybe we could do something a little bit smarter and use what are called **adaptive learning rates**

  These are learning rates that observe its landscape and adapt itself to tackle some of these challenges and maybe escape some local minima or speed up when it's on a on a local minima. This means that the learning rate, because it's adaptive it may increase or decrease depending on how large our gradient is and how fast we're learning or many other options.


In fact **these have been widely explored in deep learning** literature and heavily published on.
